```python
%matplotlib notebook

import datetime
import matplotlib.pyplot as plt
import matplotlib.dates
import sics.lakeshore.logfile as TL
```

```python
source = r'V:\data\current\2016\control.t2s4'

reader = TL.Reader(source, converter=TL.kelvin_to_celcius)
```

```python
t, v = reader.fetch_all_data()
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(t + datetime.timedelta(hours=10), v, 'k', lw=2)
ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter('%H:%M'))
ax.set_xlabel('time')
ax.set_ylabel('sample temperature /C')
```
